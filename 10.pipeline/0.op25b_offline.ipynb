```python
import glob
import shutil
vs = glob.glob('D:/climb_test/*.mp4')

for v in vs:
    video_fname = v
    out_dirname = video_fname[:-4]
    op25b_processor  = Openpose2018Offline(OFFLINE_CFG)
    ret_op25b_x2ds = op25b_processor(video_fname,out_dirname)
    a = f'{out_dirname}/op25b_x2ds.npz'
    b = f'{os.path.dirname(video_fname)}/{os.path.basename(video_fname)[:-4]}.npz'
    os.rename(a,b)
    shutil.rmtree(f'{out_dirname}/')
```

```python
play_ob25b_on_video(video_fname,ret_op25b_x2ds,interval=30)
```
