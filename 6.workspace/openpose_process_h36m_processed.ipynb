```python
import os
import sys
import numpy as np
import glob
```

```python
sys.path.append("../0.mc_utils/")
```

```python
from openpose.op_memshare import OpenPoseMemProducer
```

```python
producer = OpenPoseMemProducer(cmd_path="../2.bins/op_mc_build/bin/16_synchronous_custom_output.exe",
                               model_dir="../12.models/op25b/",
                               num_people_max=1,
                               net_resolution=384,
                               async_mode=False)
```

```python
subject="S11"
in_directories = glob.glob("E:/10.datasets/h36m-processed2/{}/*/*/*".format(subject))
out_home = "E:\\10.datasets\\op25b_h36m_jsons"

for d in in_directories:
    #step1.parse input path
    ret = d.split("\\")
    action = ret[1]
    camera = ret[-1]
    
    #step2.prepare output path
    out_path = os.path.join(out_home,subject,action,camera)
    print("\n>>>Prcessing from {} to {}".format(d,out_path))
    if not os.path.exists(out_path):
        print("Making output dir:{}".format(out_path))
        os.makedirs(out_path)
        
    #step3.start processing
    producer(image_dir=d,write_json=out_path)
```
